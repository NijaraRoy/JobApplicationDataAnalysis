```python
import pickle
import csv

class JobApplication:
    """
    This is Base class representing a job application.
    """

    def __init__(self, date, title, company, job_type, experience, salary, interview_difficulty):
        """
        Initializes a JobApplication object.

        Args:
            date (str): The application date.
            title (str): Job title.
            company (str): Company providing the job.
            job_type (str): Type of job (Stem/non-Stem).
            experience (int): Years of experience needed.
            salary (float): Salary in USD.
            interview_difficulty (float): Difficulty rating.
        """
        self.date = date
        self.title = title
        self.company = company
        self.type = job_type
        self.experience = experience
        self.salary = salary
        self.interview_difficulty = interview_difficulty
        self.unique_id = id(self)

    def __str__(self):
        """
        Returns a formatted string representation of the JobApplication object.

        Returns:
            str: Comma-separated string of all attributes.
        """
        return f"{self.unique_id},{self.date},{self.title},{self.company},{self.type},{self.experience},{self.salary},{self.interview_difficulty}"


class StemJobApplication(JobApplication):
    """
    Subclass of JobApplication representing a STEM job.
    """
    def __init__(self, date, title, company, experience, salary, interview_difficulty):
        """
        Initializes a StemJobApplication object.

        Args:
            date (str): The application date.
            title (str): Job title.
            company (str): Company providing the job.
            experience (int): Years of experience needed.
            salary (float): Salary in USD.
            interview_difficulty (float): Difficulty rating.
        """
        super().__init__(date, title, company, "Stem", experience, salary, interview_difficulty)


class NonStemJobApplication(JobApplication):
    """
    Subclass of JobApplication representing a non-STEM job.
    """
    def __init__(self, date, title, company, experience, salary, interview_difficulty):
        """
        Initializes a NonStemJobApplication object.

        Args:
            date (str): The application date.
            title (str): Job title.
            company (str): Company providing the job.
            experience (int): Years of experience needed.
            salary (float): Salary in USD.
            interview_difficulty (float): Difficulty rating.
        """
        super().__init__(date, title, company, "non-Stem", experience, salary, interview_difficulty)

#  three additional subclasses

class Apple(StemJobApplication):
    """
    Subclass of StemJobApplication representing an Apple job.
    """
    def __init__(self, date, title, experience, salary, interview_difficulty):
        """
        Initializes an Apple job application.

        Args:
            date (str): The application date.
            title (str): Job title.
            experience (int): Years of experience needed.
            salary (float): Salary in USD.
            interview_difficulty (float): Difficulty rating.
        """
        super().__init__(date, title, "Apple", experience, salary, interview_difficulty)


class Siemens(NonStemJobApplication):
    """
    Subclass of NonStemJobApplication representing a Siemens job.
    """
    def __init__(self, date, title, experience, salary, interview_difficulty):
        """
        Initializes a Siemens job application.

        Args:
            date (str): The application date.
            title (str): Job title.
            experience (int): Years of experience needed.
            salary (float): Salary in USD.
            interview_difficulty (float): Difficulty rating.
        """
        super().__init__(date, title, "Siemens", experience, salary, interview_difficulty)


class Boeing(NonStemJobApplication):
    """
    Subclass of NonStemJobApplication representing a Boeing job.
    """
    def __init__(self, date, title, experience, salary, interview_difficulty):
        """
        Initializes a Boeing job application.

        Args:
            date (str): The application date.
            title (str): Job title.
            experience (int): Years of experience needed.
            salary (float): Salary in USD.
            interview_difficulty (float): Difficulty rating.
        """
        super().__init__(date, title, "Boeing", experience, salary, interview_difficulty)

# Load the pickle file

def load_pickle_data(file_path):
    """
    Load objects from a pickle file.

    Args:
        file_path (str): Path to the pickle file.

    Returns:
        object: The deserialized object(s) from the pickle file.
    """
    with open(file_path, 'rb') as f:
        objects = pickle.load(f)
    return objects

def save_to_csv(objects, file_path):
    """
    Save a list of objects to a CSV file with the specified column headers.

    Args:
        objects (list): List of objects to save. Each object must have a valid `__str__` method.
        file_path (str): Path to save the CSV file.
    """
    with open(file_path, 'w') as f:
        f.write("unique_id,date,title,company,type,experience,salary,interview_difficulty\n")
        for obj in objects:
            f.write(str(obj) + '\n')
    print(f"CSV file generated successfully at: {file_path}")

# def save_to_csv(objects, file_path):
#     """
#     Save a list of objects to a CSV file with a specific column structure.

#     This function creates a CSV file with a header row
#     "unique_id,date,title,company,type,experience,salary,interview_difficulty".
#     Each object in the provided list must implement a `__str__` method that
#     returns a comma-separated string representation of its attributes.

#     Args:
#         objects (list): List of objects to save.
#         file_path (str): Path where the CSV file will be saved.

#     Returns:
#         None
#     """
#     with open(file_path, 'w') as f:
#         writer = csv.writer(f)
#         # Write the header row
#         writer.writerow([
#             "unique_id", "date", "title", "company",
#             "type", "experience", "salary", "interview_difficulty"
#         ])

#         # Write each object's string representation as a row
#         for obj in objects:
#             writer.writerow(str(obj).split(','))
#     print(f"CSV file generated successfully at: {file_path}")


if __name__ == "__main__":

    # Build and test OOP classes given in the instruction set
    # Create instances of all classes
    job_application = JobApplication("2023-11-17", "Generic Job", "Generic Company", "Generic Type", 5, 70000, 3.5)
    stem_job_application = StemJobApplication("2023-11-16", "Engineer", "TechCorp", 3, 85000, 4.0)
    non_stem_job_application = NonStemJobApplication("2023-11-15", "Manager", "RetailCorp", 7, 90000, 2.8)
    apple_job_application = Apple("2023-11-14", "Software Developer", 2, 120000, 4.5)
    siemens_job_application = Siemens("2023-11-13", "Electrical Engineer", 4, 95000, 3.8)
    boeing_job_application = Boeing("2023-11-12", "Analyst", 6, 87000, 3.2)

    # Print the string representation of each object
    print("JobApplication:", str(job_application))
    print("StemJobApplication:", str(stem_job_application))
    print("NonStemJobApplication:", str(non_stem_job_application))
    print("AppleJobApplication:", str(apple_job_application))
    print("SiemensJobApplication:", str(siemens_job_application))
    print("BoeingJobApplication:", str(boeing_job_application))

    # Save all objects to a CSV file
    all_objects = [
        job_application,
        stem_job_application,
        non_stem_job_application,
        apple_job_application,
        siemens_job_application,
        boeing_job_application
    ]
    save_to_csv(all_objects, 'test_job_applications.csv')

    # Load the given pickle file to retrieve the objects.
    objects = load_pickle_data('JobApplicationpkl638200306.dat')

    # Generate a csv file from the loaded objects.
    save_to_csv(objects, 'job_applications.csv')
```

## Handling Missing values

```python
# Handling Missing Values for All Columns
"""
Dynamic handling of missing values using vectorized operations in Pandas.
Strategies:
1. **Categorical Columns**:
   - Fill missing values with the mode.
2. **Numerical Columns**:
   - Fill missing values with the median or mean based on distribution characteristics.
"""

# Check for missing values
print("Checking for Missing Values:")
missing_values = df.isnull().sum()
print(missing_values)

if missing_values.sum() == 0:
    print("\nNo missing values found in the dataset.")
else:
    print("\nHandling Missing Values...")

    # Handle categorical columns using the mode
    categorical_cols = df.select_dtypes(include=['object']).columns
    df[categorical_cols] = df[categorical_cols].apply(lambda col: col.fillna(col.mode()[0]))

    # Handle numerical columns
    numerical_cols = df.select_dtypes(include=['number']).columns
    for col in numerical_cols:
      skewness = df[col].skew()
      if skewness > 1 or skewness < -1:  # High skewness
          df[col] = df[col].fillna(df[col].median())
          print(f"Filled missing values in highly skewed numerical column: {col} using median.")
      else:
          df[col] = df[col].fillna(df[col].mean())
          print(f"Filled missing values in numerical column: {col} using mean.")

    print("Missing values handled successfully.")

    # Confirm no missing values remain
    print("\nMissing Values after Cleaning:")
    print(df.isnull().sum())
```

# Explanation of handling missing data:

The dataset contains both categorical variables and numericals value.

**Handling Missing Values**:
To ensure the dataset is clean, we implemented a approach that dynamically handles missing values based on column data types. Here’s how we addressed the issue:

**Categorical Variables:**

  For fields like company, type, and outcome, missing values were replaced with the **mode** (most frequent value). This ensures the imputed values align with the most common category.

**Numerical Variables:**

For numerical fields like experience, salary, and interview_difficulty, missing values were handled dynamically by evaluating the skewness of each column. The strategy was tailored to ensure appropriate handling of symmetric and skewed distributions:

**High Skewness (greater than 1 or less than -1):**

If the data distribution showed high skewness, the median was used to fill missing values. This is because the median is unaffected by extreme values, making it more robust for heavily skewed data.

**Low to Moderate Skewness (between -1 and 1):**

For less skewed or symmetric data, the mean was used. The mean is a reliable measure of central tendency for such distributions.

**About Skewness:**
**Definition**: Skewness quantifies the degree of asymmetry in a data distribution.

**Low to Moderate Skewness:** Indicates the data is approximately symmetric, and mean can be used reliably.

**High Skewness:** Suggests a long tail or outliers, making the median a better choice for imputation.

**Handling Process:**
For each numerical column:
Skewness was calculated using the .skew() method to determine the symmetry of the data.

**Based on the skewness:**
Columns with high skewness were imputed with the median.
Columns with low to moderate skewness were imputed with the mean.
This ensured that imputation decisions were data-driven and minimized the impact of outliers on skewed data.

To make the process efficient, we used lambda functions instead of explicit loops.

For categorical columns, a lambda function fills missing values with the mode dynamically.
For numerical columns, a lambda function calculates skewness and applies either the mean or median based on the result.


The above code section automatically detects missing values in all columns,
Dynamically applies the appropriate imputation method (mode, mean, or median) based on the column type and data distribution.

**Outcome:**
After running the function:

Missing values in categorical variables were replaced with the mode.
Missing values in numerical variables were handled using skewness to decide between mean and median.
The dataset is now clean and ready for further analysis.

This approach ensures that missing values are handled efficiently and dynamically for any dataset structure.

# Pie Charts:
Pie charts are best used to show proportions or percentages of a whole. They help illustrate how different parts contribute to the total and are ideal for categorical data.
```python
# Generate Pie chart for job application for each company

# Calculate the number of applications for each company
applications_by_company = df['company'].value_counts()

# Define labels and sizes
labels = applications_by_company.index
sizes = applications_by_company.values

# Define colors for each segment
colors = ['red', 'blue', 'green']

# Explode the largest segment for emphasis
explode = [0.1 if i == 0 else 0 for i in range(len(labels))]

# Create the pie chart
plt.figure(figsize=(8, 8))
plt.pie(
    sizes,
    labels=labels,
    colors=colors,
    explode=explode,
    autopct='%1.1f%%',
    startangle=140,
    textprops={'fontsize': 10}
)
plt.title('Job Applications by Company', fontsize=16, fontweight='bold')
plt.show()

```

# 3. Line Plots
Line plots are best used for visualizing trends or relationships in data over a continuous variable. They connect data points with a line to show how a variable changes over time or other continuous dimensions.Our goal here is to understand how average salaries have changed over months, indicating potential market trends or seasonal effects.


```python
import matplotlib.pyplot as plt
import pandas as pd

# Ensure the 'date' column is in datetime format
df['date'] = pd.to_datetime(df['date'])

# Group data by month and calculate the average salary
salary_over_time = df.groupby(df['date'].dt.to_period('M'))['salary'].mean()

# Plot the trend
plt.figure(figsize=(10, 6))
salary_over_time.plot(marker="*", markersize=15, markerfacecolor='red', linestyle='-', color='blue', grid=True)
plt.title('Average Salary Over Time', fontsize=16, fontweight='bold')
plt.xlabel('Time (Monthly)', fontsize=12)
plt.ylabel('Average Salary', fontsize=12)
plt.xticks(rotation=45, fontsize=10)
plt.yticks(fontsize=10)
plt.tight_layout()
plt.show()
```

```python
# Plot 2: Salary Trend by Company
# 2. Average Salary Over Time by Company
company_salary_trend = df.groupby([df['month_year'], 'company'])['salary'].mean().unstack()
plt.figure(figsize=(10, 6))
company_salary_trend.plot(marker='o')
plt.title('Average Salary Over Time by Company', fontsize=16, fontweight='bold')
plt.xlabel('Time (Monthly)', fontsize=12)
plt.ylabel('Average Salary', fontsize=12)
plt.grid(alpha=0.5)
plt.legend(title='Company', fontsize=10)
plt.tight_layout()
plt.show()

```

# 7. Heat Maps:

Heat maps are best suited for visualizing relationships, patterns, or intensity in data that is represented in a matrix-like structure. They work well with numerical data that can be compared across two dimensions.In our dataset, a Correlation Heat Map can be used to examine the relationships between salary, experience, and interview difficulty.

```python
# Correlation Heatmap
plt.figure(figsize=(8, 6))  # Set the figure size for the heatmap
sns.heatmap(
    df[['experience', 'salary', 'interview_difficulty']].corr(),  # Compute correlations between the selected columns
    annot=True,  # Display correlation values on the heatmap
    cmap='coolwarm',  # Use the 'coolwarm' colormap for visual appeal
    fmt=".2f",
    linewidths=0.5
)
plt.title('Correlation Heatmap', fontsize=16, fontweight='bold')
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.show()  # Display the plot

```

# Data Trends Based on Correlation Heatmap:



**Weak Correlation Between Variables:**

The heatmap indicates that there is almost no correlation between experience, salary, and interview_difficulty.
All correlation coefficients are close to zero, meaning changes in one variable have little to no impact on the others.

**No Direct Relationship:**

The lack of significant positive or negative correlations implies that factors like interview_difficulty or experience are not strongly tied to variations in salary in this dataset.

**Diagonal Dominance:**

The diagonal values show a perfect correlation (1.00) as these are the same variables compared with themselves, which is expected.

**Overall Insight:**

The heatmap reveals that the selected variables (experience, salary, and interview difficulty) are mostly independent of each other in this dataset.
